```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv("athlete_events.csv")
df2 = pd.read_csv("noc_regions.csv")
```

```python
df1.head()
```

```python
df2.head()
```

```python
# join 2 dataset on NOC
df = df1.merge(df2, on='NOC', how='left')
```

```python
df.sample(5)
```

```python
df.info()
```

```python
# rename the column name consitent
df.rename(columns={"region":"Region","notes":"Notes"},inplace=True)
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
#India details
df[df["Team"]=="India"]
```

```python
# Indian teams with gold medals
df.query('Team == "India" and Medal == "Gold"')
```

```python
# top 10 Region participating

df.groupby("Region")["Region"].count().sort_values(ascending=False).head(10)
```

```python
# let's plot a bar for top 10 participating region

df["Region"].value_counts().sort_values(ascending=False)[:10].plot(kind="bar",figsize=(10,5),color='red')
plt.title('Top 10 participating region',fontsize=14)
plt.ylabel('No of participants')
plt.xlabel('Region Name')
plt.show()
```

```python
# Age Distribution 
plt.figure(figsize=(12,6))
plt.title('Age Distribution')
plt.hist(df["Age"],bins=np.arange(10,80,2),color='red')
```

```python
# winter olympics game
df[df["Season"]=="Winter"].Sport.unique()
```

```python
# Summer olympics game
df[df["Season"]=="Summer"].Sport.unique()
```

```python
# male and female participants
df["Sex"].value_counts()
```

```python
# let's Plot a bar plot for male and female athletes
df["Sex"].value_counts().plot(kind='bar')
```

```python
# Total medal
df["Medal"].value_counts()
```

```python
pd.crosstab(df['NOC'],df['Medal']).sort_values(by='Gold',ascending=False)[:10]
```

```python
pd.crosstab(df['NOC'],df['Medal']).sort_values(by='Gold',ascending=False)[:10].plot(kind='bar',figsize=(11,5))
```

```python
# most number of medal by Player
df.groupby('Name')["Medal"].count().sort_values(ascending=False).head(10)
```

```python
df.groupby('Name')["Medal"].count().sort_values(ascending=False).head(10).plot(kind="bar",figsize=(11,5),color='red')
plt.xlabel('Name of Player')
plt.ylabel('Total No of Medals')
```
